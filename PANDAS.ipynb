#Time series


```python
#what is time series data?
"""Time series data is a sequence of data points in cronological order that is used by
business to analyze the past data and make the future predictions"""
#ex:- weather forcast
#No of taxi rides in a month
#predicting a company stock price for the next day
```

```python
#components of the time series
#it is consist of 4 components
#1. Trend Component:- This is a varaiation that moves up or down ina predictable pattern over long period
#2.sesonality component:- The variation  that is regualr and periodic and repeats itself over a specific period
    #ex:- such as days, weeks, or the months, seasons
#3. cyclic component:- is the variation that corresponds with business or economic,  "boom-bust"  cycles it folows their own peculiar cycles
#4.Randomn component:- irregular component
```

#pivot table


```python
df = pd.DataFrame(
    {
        "A":['one','one','two','three'] *3,
        "B":['A','B','C']*4,
        "C":['food','food','food','drink','drink','drink']*2,
        'D':np.random.randint(0,12, size = 12),
        'E':np.random.randint(0,12,size= 12),

    }
)
df
```

```python
#pd.pivot_table()
pd.pivot_table(df,values='D',index=['A','B'],columns=['C'])
```

#group by

-- splitting

-- applying

-- combining


```python
df = pd.DataFrame(
    {
        "A":['one','one','two','three'] *3,
        "B":['A','B','C']*4,
        "C":['food','food','food','drink','drink','drink']*2,
        'D':np.random.randn(12),
        'E':np.random.randn(12)

    }
)
df
```

```python
#grouby function is applicable with the aggregate function
#min, max, sum(),avg, count,median, var,std
```

```python
#print the sum of food and drink  for "D" and "E"
df.groupby("C")[["D","E"]].std()
```

```python
df.groupby(['A','C']).sum()
```
